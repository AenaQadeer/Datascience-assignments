```python
#Question 1
import pandas as pd
```

```python
pf=pd.read_csv("netflix_data.csv")
```

```python
pf.head()
```

```python
import numpy as np
```

```python
pf["cast"]=pf["cast"].replace(np.NaN,"null")
```

```python
pf.head()
```

```python
def num_of_cast(total):
    total_values=total.split(",")
    return len(total_values)
```

```python
pf["num_of_cast"]=pf.loc[:,"cast"].apply(num_of_cast)
```
